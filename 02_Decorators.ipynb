# Decorators
to modify behaviour of existing function

1. Input is any function
2. Output is a modifed function -> Wrapper

# Write a decorator to welcome and thank the user

```python
def welcome(func):
    def wrapper(*args, **kwargs):
        print("Hello User!")
        res = func(*args, **kwargs)
        print(f"Function Results : {res}")
        print("Thank you")
        return res
    return wrapper
```

# If you want to apply decorator @decorator

```python
@welcome
def simple_interest(p: float, n: int, r: float) -> tuple:
    i = (p * n * r) / 100
    a = p + i
    return i, a
```

```python
i1, a1 = simple_interest(p=50_000, n=5, r=7.1)
```

# Write hypotenuse function with welcome decorator

```python
import math

@welcome
def hypotenuse(a: float | int, b: float | int) -> float:
    h = math.sqrt(a**2 + b**2)
    return h
```

```python
h1 = hypotenuse(a=3, b=4)
```

```python
h2 = hypotenuse(5, 6)
```

# Vowel counter welcome decorater

```python
@welcome
def vowel_counter(word: str):
    word = word.lower()
    count = 0
    for i in word:
        if i in "aeiou":
            count = count + 1
    return count
```

```python
c1 = vowel_counter("ETLHive")
```

```python
c2 = vowel_counter("Utkarsh")
```

```python
c3 = vowel_counter("Aditi")
```
