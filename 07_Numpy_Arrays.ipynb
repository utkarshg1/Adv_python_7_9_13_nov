# Multiplication in arrays vs list

```python
a = [1, 2, 3, 4, 5]
type(a)
```

```python
# In list multiplication means repeating the list n times
a * 3
```

```python
a * 1.5
```

```python
# Multiplying by 3 inside a list
b = []

for i in a:
    print(f"Multiplication of {i} by 3 is {i * 3}")
    b.append(i * 3)

print(b)
```

```python
# Numpy arrays support broadcasting
a1 = np.array(a)
print(a1, type(a1))
```

```python
b1 = 3 * a1
print(b1)
```

```python
1.5 * a1 + 3
```

```python
# Square
a1 ** 2
```

```python
# Cube
a1 ** 3
```

```python
# Square root
a1 ** (1/2)
```

```python
# Cube root
a1 ** (1/3)
```

# Taking product in list vs arrays

```python
c = [4, 5, 6, 7, 8, 9]
print(c, type(c))

# Initialize product by 1
p = 1
for i in c:
    p = p * i

# Print the product
print(f"Product of elements inside list {c} is {p}")
```

```python
# Numpy array product
c1 = np.array(c)
print(c1, type(c1))
p1 = c1.prod()
print(f"Product of array {c1} is {p1}")
```
